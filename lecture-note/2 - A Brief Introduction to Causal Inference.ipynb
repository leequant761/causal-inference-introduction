# 1. What are potential outcomes?

* 약을 먹을 때의 나, 약을 안먹을 때의 나를 동시에 가르키기 위한 개념

* $\left.Y_{i}\right|_{\mathrm{do}(T=1)} \triangleq Y_{i}(1)$
  * 약을 먹었을 때의 subject $i$의 outcome variable
* $\left.Y_{i}\right|_{\mathrm{do}(T=0)} \triangleq Y_{i}(0)$
  * 약을 안먹었을 때의 subject $i$의 outcome variable
* $Y_{i}(1) - Y_{i}(0)$
  * Individual Treatment Effect (ITE)

# 2. The fundamental problem of causal inference

* $i$ 사람을 갖고서 $T=t$ 를 투여하게 된다면 $Y_i(t)$를 관측하게 됨
  * 하지만, $Y_i(t^\prime)$은 영영 관측하지 못하게 됨
  * 이는 결측치가 포함된 자료로 해석할 수 있음

# 3. Getting around the fundamental problem of causal inference

* $\mathbb E [Y_{i}(1) - Y_{i}(0)] = \mathbb E [Y(1)] - \mathbb E [Y(0)]$
  * Average Treatment Effect (ATE)
  * $\neq \mathbb E [Y \mid T=1] - \mathbb E [Y \mid T=0]$
    * confounder가 존재하지 않아야 등호가 성립

## 3.1 Ignorability (Exchangeability)

* $Y(1) , Y(0) \perp T $

$$
\begin{aligned}
\mathbb E [Y_{i}(1) - Y_{i}(0)] &= \mathbb E [Y(1)] - \mathbb E [Y(0)]
\\
&= \mathbb E [Y(1) \mid T=1] - \mathbb E [Y(0) \mid T=0] \quad \because \text{ignorability}
\\
&= \mathbb E [Y \mid T=1] - \mathbb E [Y \mid T=0] \quad \because \text{ignorability}
\end{aligned}
$$

* Exchangeability라고 불리는 이유
  * 실험군과 대조군을 실수로 헷갈려서 반대로 처리해도
  * 즉, 바꿔도 상관없음에서 유래

### 3.1.1 Conditional Exchangeability

<img src="2 - Potential Outcomes.assets\image-20220816121631411.png" alt="image-20220816121631411">

* $Y(1) , Y(0) \perp T \mid X$

$$
\begin{aligned}
\mathbb{E}[Y(1)-Y(0) \mid X] &=\mathbb{E}[Y(1) \mid X]-\mathbb{E}[Y(0) \mid X] \\
&=\mathbb{E}[Y(1) \mid T=1, X]-\mathbb{E}[Y(0) \mid T=0, X] \\
&=\mathbb{E}[Y \mid T=1, X]-\mathbb{E}[Y \mid T=0, X]
\end{aligned}
$$

* Observational data에서 exchangeability는 비현실적 가정임

## 3.2 Identifiability

* Causal quantity가 Statistical quantity로 계산될 수 있음을 의미
* $\text{do}$-operator 없이 $P(v)$ 만으로 표현가능 여부

## 3.3 Assumptions for valid inference

* Unconfoundness
  * $Y(1) , Y(0) \perp T \mid X$
    * for some observed covariates $X$
* Positivity
  * $0 < P(T=1 \mid X=x) < 1$
    * for all values of covariates $X$
  * 위배될 경우, $\mathbb{E}_X[\mathbb{E}[Y \mid T=1, X]-\mathbb{E}[Y \mid T=0, X]]$이 정의되지 않음
  * covariate이 고차원일 수록 extrapolation을 요구

* SUTVA
  * stable unit-treatment value assumption
  * 이는 2개의 가정의 결합을 의미
  * No interference
    * 다른 이의 treatment가 나의 potential outcome에 영향주지 않음
    * 이것은 전염성(contagious)있는 agent를 다룰 때, individual outcome이 social interaction에 의해서 영향을 받기 때문에 깨짐
  * Consistency
    * $T=t$가 $Y(t)=Y$를 의미
    * 수술을 담당하는 의사가 다르다면 같은 수술을 받았지만
      * $Y_i(1), Y_{i+1}(1)$ 의 구분이 있어야 함

---

$$
\begin{aligned}

\mathbb{E}[Y(1)-Y(0)] &=\mathbb{E}[Y(1)]-\mathbb{E}[Y(0)] 
\\&\quad \text { (linearity of expectation) } 
\\
&= \mathbb{E}_{X}[\mathbb{E}[Y(1) \mid X]-\mathbb{E}[Y(0) \mid X]] 
\\& \quad \text { (law of iterated expectations) } \\
&=\mathbb{E}_{X}[\mathbb{E}[Y(1) \mid T=1, X]-\mathbb{E}[Y(0) \mid T=0, X]] 
\\&\quad \text { (unconfoundedness and positivity) } 
\\
& =\mathbb{E}_{X}[\mathbb{E}[Y \mid T=1, X]-\mathbb{E}[Y \mid T=0, X]] \\&\quad \text { (consistency) } 
\end{aligned}
$$
