# 1. The $do$-operator

<img src="4 - Causal Models.assets\image-20220818151644057.png" alt="image-20220818151644057">

* 기존 시스템에서 subpopulation을 잡는 것이 conditioning
* Population에 treatment를 일괄적부여(개입)하는 것이 intervening

## Notation

* Interventional distribution
  * $do$-operator가 포함된 확률분포
  * $P(Y(t) = y) :=: P(Y=y \mid do(T=t)) :=: P(y \mid do(t))$
* ATE
  * $\mathbb E[Y(1)] - \mathbb E[Y(0)] := \mathbb E[Y \mid do(T=1)] - \mathbb E[Y \mid do(T=0)]$

# 2. Main assumption : modularity

<img src="4 - Causal Models.assets\image-20220818153423970.png" alt="image-20220818153423970">

* $x_i$가 아닌 모든 변수에 $do$를 가하여도 $P(x_i \mid pa_i)$는 변하지 않는다.
* $x_i$에 $do$를 가하면 $P(x_i \mid pa_i)$는 1또는 0이다.
* invariance, independent mechanism 등으로도 불린다.

## Truncated Factorization

$$
P(x_i, \ldots, x_n \mid do(S=s)) = \prod_{i \not\in S} P(x_i \mid pa_i) \quad \text{or} \quad 0
$$

### Example : Association is not causation

<img src="4 - Causal Models.assets\image-20220818163010217.png" alt="image-20220818163010217">

$$
\begin{aligned}
P(y, t, x) &= P(x) P(t \mid x) P(y\mid t, x)
\\
P(y, x \mid do(t)) &= P(x) P(y \mid t, x)
\\
P(y \mid do(t)) &= \sum_xP(x) P(y\mid t, x)
\\
\\
P(y\mid t) &= \sum _x P(x \mid t) P(y\mid t, x) 
\end{aligned}
$$

# 3. Backdoor adjustment

<img src="4 - Causal Models.assets\image-20220818163552252.png" alt="image-20220818163552252">

* $T$에서 $Y$로 가는 backdoor path를 막음으로서 causal effect에 집중

## 3.1 Backdoor criterion

* $W$는 $T\rightarrow Y$에 관해 backdoor criterion을 만족한다. if
  * $W$ blocks all backdoor paths from $T$ to $Y$
  * $W$ does not contain any descendants of $T$

## 3.2 Backdoor adjustment

* $W$가 backdoor criterion을 만족할 경우

$$
\begin{aligned}
P(y \mid do(t)) &= \sum_w P(y \mid do(t), w) P(w \mid do(t))
\\
&= \sum_w P(y \mid t, w) P(w \mid do(t)) 
\\
&\quad \because \text{modularity}
\\
&= \sum_w P(y \mid t, w) P(w) 
\\
&\quad \because \text{modularity}


\end{aligned}
$$

# 4. Structural causal models

## 4.1 Structural Equations

<img src="4 - Causal Models.assets\image-20220818164917231.png" alt="image-20220818164917231">

* $B := f(A, U)$
  * $A$ is a cause of $B$

### Causal mechanisms (modularity)

<img src="4 - Causal Models.assets\image-20220818153423970.png" alt="image-20220818153423970">

* $P(x_i \mid pa_i)$
* $X_i := f(PA_i, U)$
* 둘 중 어느 것을 명시하든 치환이 가능함

### 4.2 Structural causal models (SCMs)

<img src="4 - Causal Models.assets\image-20220818165429223.png" alt="image-20220818165429223">
$$
M : \quad \begin{aligned}
B &:= f_B(A, U_B)
\\
C &:= f_C(A, B, U_C)
\\
D &:= f_D(A, C, U_D)
\end{aligned}
$$

* $(V, U, F)$ 에 대해 명시하는 것
* $V$ is a set of endogenous variables
* $U$ is a set of exogenous variables
* $F$ is a set of functions

### 4.2.1 Interventions

<img src="4 - Causal Models.assets\image-20220818165907283.png" alt="image-20220818165907283">

$$
M_b : \quad \begin{aligned}
B &:= b
\\
C &:= f_C(A, B, U_C)
\\
D &:= f_D(A, C, U_D)
\end{aligned}
$$
