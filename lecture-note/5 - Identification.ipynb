# 1. Frontdoor adjustment

<img src="5 - Identification.assets\image-20220818220757967.png" alt="image-20220818220757967">

1. Identify CE of $T$ on $M$
   * $P(m \mid do(t)) = P(m\mid t)$
   * Backdoor adjustment with $\emptyset$
2. Identify CE of $M$ on $Y$
   * $P(y \mid do(m)) = \sum_t P(y\mid m, t) P(t)$
   * Backdoor adjustment with $T$
3. Identify CE of $T$ on $Y$
   * $P(y \mid do(t)) = \sum_m P(m \mid do(t)) P(y\mid do(m))$
   * 직감
     * $T \rightarrow M$ 으로 가는 causal effect
     * $M \rightarrow Y$ 으로 가는 causal effect
     * 를 각각 아니깐 잘 이어붙이면 되지 않을까?

---

$$
\begin{aligned}
P(w, t, m, y) &= P(w)P(t\mid w)P(m \mid t) P(y\mid w, m)
\\
P(w, m, y \mid do(t)) &= P(w)P(m \mid t) P(y\mid w, m)
\\
\\
P(y \mid do(t)) &= \sum_{m}P(m \mid t)\sum_{w} P(y\mid w, m) P(w)
\\
& \quad \because \text{marginalization}
\\
&= \sum_{m}P(m \mid t)\sum_{w} P(y\mid w, m) \sum_{t^\prime}P(w\mid t^\prime)P(t^\prime)
\\
&= \sum_{m}P(m \mid t) \sum_{t^\prime} P(t^\prime) \sum_{w} P(y\mid w, m)P(w\mid t^\prime, m)
\\
&= \sum_{m}P(m \mid t) \sum_{t^\prime} P(t^\prime) \sum_{w} P(y\mid w, t^\prime, m)P(w\mid t^\prime, m)
\\
&\quad \because \text{modularity}
\\
&= \sum_{m}P(m \mid t) \sum_{t^\prime} P(t^\prime) \sum_{w} P(y, w \mid t^\prime, m)
\\
&= \sum_{m}P(m \mid t) \sum_{t^\prime} P(t^\prime)P(y \mid t^\prime, m)
\\
&= \sum_m P(m \mid do(t)) P(y\mid do(m))
\end{aligned}
$$

# 2. Pearl's $do$-calculus

* causal graph에 기반해 causal quantity를 identify하는 tool

## Notation

* $G_{\underline T}$
* $G_{\overline T}$
* $\perp_G$
* $\mathbf{Z(W)}$:  $G_{\overline {\mathbf T}}$에서 $\mathbf W$의 ancestor가 아닌 $\mathbf Z$의 subset

## Definition

* Rule1 :  D-separation의 일반화
  * 만약 $\mathbf Y \perp _{G_{\overline{\mathbf T}}} \mathbf Z \mid \mathbf {T, W}$ 라면,
    * $p^{\mathfrak{C} ; d o(\mathbf{T}:=\mathbf{t})}(\mathbf{y} \mid \mathbf{z}, \mathbf{w})=p^{\mathfrak{C} ; d o(\mathbf{T}:=\mathbf{t})}(\mathbf{y} \mid \mathbf{w})$
  * $\mathbf T = \emptyset$일 경우 D-separation이다.
* Rule2 : Backdoor adjustment의 일반화
  * 만약 $\mathbf Y \perp _{G _{\overline{\mathbf T}, \underline{\mathbf Z}}}\mathbf Z \mid \mathbf {T}, \mathbf{W}$ 라면,
    * $p^{\mathfrak{C} ; d o(\mathbf{T}:=\mathbf{t}, \mathbf{Z}=\mathbf{z})}(\mathbf{y} \mid \mathbf{w})=p^{\mathfrak{C} ; d o(\mathbf{T}:=\mathbf{t})}(\mathbf{y} \mid \mathbf{z}, \mathbf{w})$
  * $\mathbf T = \emptyset$일 경우 Backdoor adjustment이다.
* Rule3 : Insertion/deletion of actions
  * 만약 $\mathbf Y \perp _{G_{\overline{\mathbf{T}}, \overline {\mathbf{Z(W)}}}}\mathbf Z \mid \mathbf {T, W}$라면,
    * $p^{\mathfrak{C} ; d o(\mathbf{T}:=\mathbf{t}, \mathbf{Z}=\mathbf{z})}(\mathbf{y} \mid \mathbf{w})=p^{\mathfrak{C} ; d o(\mathbf{T}:=\mathbf{t})}(\mathbf{y} \mid \mathbf{w})$
  * 해석 :
    * $\mathbf Z \setminus \mathbf Z(\mathbf W)$와 $\mathbf Y$ 사이의 열린 모든 정방향으로 시작하는 path는 $\mathbf W$에 의해 block되었고
    * $\mathbf Z(\mathbf W)$에서 $\mathbf Y$로의 causal path는 존재하지 않으면,
    * intervention을 지워도 됨

## Example

<img src="5 - Identification.assets\image-20220818220757967.png" alt="image-20220818220757967">

$$
\begin{aligned}
P(y \mid do(t)) &= \sum_m P(y \mid do(t), m) P(m \mid do(t))
\\
&= \sum_m P(y \mid do(t), m) P(m \mid t) \quad \text{Rule2} \because M \perp _{G_{\underline T}} T
\\
&= \sum_m P(y \mid do(t), do(m)) P(m \mid t) \quad \text{Rule2} \because M \perp _{G_{\overline T \underline M}} M \mid T
\\
&= \sum_m P(y \mid do(m)) P(m \mid t) \quad \text{Rule3}
\\
&= \sum_m P(m \mid t) P(y \mid do(m)) 
\\
&= \sum_m P(m \mid t) \sum_{t^\prime}P(y \mid do(m), t^\prime)P(t^\prime \mid do(m))
\\
&= \sum_m P(m \mid t) \sum_{t^\prime}P(y \mid m, t^\prime)P(t^\prime \mid do(m)) \quad \text{Rule2}
\\
&= \sum_m P(m \mid t) \sum_{t^\prime}P(y \mid m, t^\prime)P(t^\prime) \quad \text{Rule3}
\end{aligned}
$$
